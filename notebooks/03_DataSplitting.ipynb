1. Imports and Loading

```python
import pandas as pd

from sklearn.model_selection import train_test_split 

df = pd.read_csv(
    "../data/processed/metabric_processed.csv"
)

df.head()
```

2. Separate Features and Targets

```python
X = df.drop(
    columns=[
        "overall_survival_months",
        "overall_survival"
    ]
)

y_time = df["overall_survival_months"]

y_event = df["overall_survival"]
```

Survival analysis differs from conventional classification problems by requiring two target variables:

- **Overall survival time**, representing the observed survival duration in months.
- **Overall survival event**, indicating whether the event (death) occurred (`1`) or whether the observation was censored (`0`).

The remaining variables constitute the feature matrix used for model development.

3. Train/Test Split

```python
X_train, X_test, y_time_train, y_time_test, y_event_train, y_event_test = train_test_split(
    X,
    y_time,
    y_event,
    test_size=0.2,
    random_state=42,
    stratify=y_event
)
```

```python
print(X_train.shape)
print(X_test.shape)

print()

print(y_event_train.value_counts(normalize=True))

print()

print(y_event_test.value_counts(normalize=True))
```


The processed dataset was divided into training (80%) and testing (20%) subsets.

A **stratified split** was performed using the survival event indicator to preserve the proportion of observed events and censored patients across both subsets. Maintaining a similar class distribution helps ensure that model evaluation remains representative of the underlying cohort.

4. Saving Split Dataset

```python
X_train.to_csv(
    "../data/processed/train_features.csv",
    index=False
)

X_test.to_csv(
    "../data/processed/test_features.csv",
    index=False
)

y_time_train.to_csv(
    "../data/processed/train_time.csv",
    index=False
)

y_time_test.to_csv(
    "../data/processed/test_time.csv",
    index=False
)

y_event_train.to_csv(
    "../data/processed/train_event.csv",
    index=False
)

y_event_test.to_csv(
    "../data/processed/test_event.csv",
    index=False
)
```
