# A. Random Survival Forest Model Performance

## Overview

A Random Survival Forest (RSF) model was developed to predict breast cancer survival outcomes using clinical and molecular features from the METABRIC dataset.

Survival analysis differs from traditional classification because it predicts **time-to-event outcomes** while accounting for censored observations. In cancer datasets, some patients remain alive at the end of follow-up, meaning their exact survival time is unknown. Survival models incorporate these patients rather than excluding them from analysis.

The RSF model combines multiple survival decision trees to estimate individualized survival probabilities and patient-specific risk scores.


# Data and Model Configuration

The final processed dataset contained:

* Training samples: **1523 patients**
* Testing samples: **381 patients**
* Final input features: **722 clinical and molecular variables** (orginally 724, but leakages from 2 features were fixed)

During model development, leakage analysis was performed to ensure that variables directly representing patient outcomes were not used as predictors.Outcome-derived variables, including survival status indicators such as death-related fields, were removed before final evaluation.
This ensures that model performance reflects genuine predictive ability from available patient characteristics rather than direct access to the outcome being predicted.

---

# Concordance Index (C-index)
The concordance index measures how accurately a survival model ranks patients according to their predicted risk. It evaluates whether patients predicted to have higher risk experience the event earlier than patients predicted to have lower risk.

The C-index ranges from:

* **0.5** -> random prediction
* **1.0** -> perfect discrimination
* **less than 0.5** ->  worse than random

Unlike accuracy in classification problems, the C-index evaluates ordering of survival risk while accounting for censored observations.

---

## Model Result

RSF achieved:

```
C-index = [0.716]
```

This demonstrates the model's ability to distinguish between patients with different survival outcomes. A value above 0.5 indicates that the model learned meaningful patterns associated with patient risk, while higher values indicate stronger separation between high-risk and low-risk individuals.

---

# B. Integrated Brier Score (IBS)

The Integrated Brier Score evaluates the accuracy of predicted survival probabilities over the entire follow-up period. At each time point, the model's predicted survival probability is compared with the observed patient outcome. IBS measures both:

* calibration: whether predicted probabilities match actual data
* accuracy: whether predictions are close to observed outcomes

Interpretation:

* **0 = perfect prediction**
* Lower values indicate better survival probability estimates

---

## Model Result

RSF achieved:

```
Integrated Brier Score = 0.0998
```

The low IBS indicates that the model produced well-calibrated survival predictions over time. Together with the C-index, this suggests that the model not only ranked patients according to risk but also generated reliable individualized survival probability estimates.

# D. Survival Curve and Risk Score Analysis

## Predicted Survival Curves
A survival curve represents the probability that a patient remains alive beyond a given time point. The x-axis represents time in months, while the y-axis represents predicted survival probability. A curve begins near 1.0 because all patients are alive at baseline. Over time, the probability decreases as the likelihood of experiencing the event increases.

---

## Interpretation of RSF Survival Curve Result

RSF generated individualized survival curves for patients in the test cohort. Differences between curves demonstrate that the model identified distinct survival profiles:

* Curves with slower decline represent patients predicted to have better survival outcomes.
* Curves with faster decline represent patients predicted to have higher mortality risk.

The variation between curves indicates that the model was able to distinguish between patients with different predicted risk trajectories. Rather than producing a single prediction for the entire population, the RSF provides personalized survival estimates based on each patient's clinical and molecular profile.

---

# E. Predicted Risk Score Distribution

The RSF risk score summarizes the overall predicted survival risk for each patient.
A higher risk score indicates:
* greater predicted mortality risk
* shorter expected survival
A lower risk score indicates:
* lower predicted risk
* more favorable expected survival

The risk score provides a continuous measure of patient risk rather than dividing patients into only high-risk and low-risk groups.

---

## Interpretation

The distribution of predicted risk scores demonstrates variation across the patient population. A broad distribution suggests that the model identified meaningful differences between patients based on their molecular and clinical characteristics. Patients with the highest predicted risk scores represent individuals whose feature profiles resemble patterns associated with poorer survival outcomes in the training data.

# F. Conclusion

The survival curves and risk score visualization provide two perspectives on RSF predictions:

Together, these visualizations demonstrate that the model generates individualized survival predictions from biological and clinical information.

# H. Limitations

Although the Random Survival Forest demonstrated strong predictive performance, several limitations should be considered.

## Dataset limitations

The model was trained and evaluated using the METABRIC breast cancer cohort. Since patient populations, treatment strategies, and clinical practices vary across institutions, external validation on independent cohorts would be required before clinical application. (only ~1.5k patients)

## Feature interpretation limitations

Permutation importance identifies features that contribute to prediction but does not establish causal relationships between genes and survival outcomes. Some important features may act as biomarkers rather than direct drivers of disease progression.

## Model limitations

RSFs provide strong predictive capability but have limited interpretability compared with simpler statistical survival models like Cox proportional hazards models. Therefore, model predictions should be interpreted as "risk estimates" rather than definitive clinical predictions.
